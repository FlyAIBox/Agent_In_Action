````python
# å¯¼å…¥å¿…è¦çš„æ¨¡å—
import os, getpass

def _set_env(var: str):
    """
    è®¾ç½®ç¯å¢ƒå˜é‡çš„è¾…åŠ©å‡½æ•°

    å‚æ•°:
        var (str): è¦è®¾ç½®çš„ç¯å¢ƒå˜é‡åç§°

    åŠŸèƒ½:
        - æ£€æŸ¥ç¯å¢ƒå˜é‡æ˜¯å¦å·²å­˜åœ¨
        - å¦‚æœä¸å­˜åœ¨ï¼Œåˆ™æç¤ºç”¨æˆ·è¾“å…¥å¹¶è®¾ç½®
    """
    if not os.environ.get(var):  # æ£€æŸ¥ç¯å¢ƒå˜é‡æ˜¯å¦å·²è®¾ç½®
        os.environ[var] = getpass.getpass(f"{var}: ")  # å®‰å…¨åœ°è·å–ç”¨æˆ·è¾“å…¥

# è®¾ç½® OpenAI API å¯†é’¥
# è¿™æ˜¯ä½¿ç”¨ OpenAI æ¨¡å‹æ‰€å¿…éœ€çš„
_set_env("OPENAI_API_KEY")
# è®¾ç½® OpenAI APIä»£ç†åœ°å€ (ä¾‹å¦‚ï¼šhttps://api.apiyi.com/v1ï¼‰
_set_env("OPENAI_BASE_URL")
````

````python
# ===== å¯¼å…¥å¿…è¦çš„åº“ =====
import json  # ç”¨äºJSONæ•°æ®å¤„ç†
import os    # ç”¨äºæ“ä½œç³»ç»Ÿç›¸å…³æ“ä½œ
import sys   # ç”¨äºç³»ç»Ÿç›¸å…³æ“ä½œ
from typing import List, Dict   # ç±»å‹æç¤ºï¼Œæé«˜ä»£ç å¯è¯»æ€§
from openai import OpenAI       # ç”¨äºè°ƒç”¨OpenAI API
# ===== æ ¸å¿ƒå·¥å…·å‡½æ•°å®šä¹‰ =====

def extract_markdown_block(response: str, block_type: str = "json") -> str:
    """
    ä»LLMå“åº”ä¸­æå–ä»£ç å—å†…å®¹

    å‚æ•°:
        response: LLMçš„åŸå§‹å“åº”æ–‡æœ¬
        block_type: è¦æå–çš„ä»£ç å—ç±»å‹ï¼Œé»˜è®¤ä¸º"json"

    è¿”å›:
        æå–å‡ºçš„ä»£ç å—å†…å®¹
    """
    # æ£€æŸ¥å“åº”ä¸­æ˜¯å¦åŒ…å«ä»£ç å—æ ‡è®°
    if not '```' in response:
        return response

    # åˆ†å‰²å“åº”å¹¶æå–ç¬¬ä¸€ä¸ªä»£ç å—
    code_block = response.split('```')[1].strip()

    # å¦‚æœä»£ç å—ä»¥æŒ‡å®šç±»å‹å¼€å¤´ï¼Œåˆ™ç§»é™¤ç±»å‹æ ‡è¯†
    if code_block.startswith(block_type):
        code_block = code_block[len(block_type):].strip()

    return code_block

def generate_response(messages: List[Dict]) -> str:
    """
    è°ƒç”¨LLMç”Ÿæˆå“åº”

    å‚æ•°:
        messages: æ¶ˆæ¯åˆ—è¡¨ï¼ŒåŒ…å«ç³»ç»Ÿæç¤ºå’Œå¯¹è¯å†å²

    è¿”å›:
        LLMç”Ÿæˆçš„å“åº”æ–‡æœ¬
    """
    # ä½¿ç”¨LiteLLMè°ƒç”¨OpenAI GPT-4oæ¨¡å‹

    client=OpenAI(
        base_url=os.environ['OPENAI_BASE_URL'],
        api_key=os.environ['OPENAI_API_KEY']
    )
    response = client.chat.completions.create(
        model="gpt-4o",  # æŒ‡å®šä½¿ç”¨çš„æ¨¡å‹
        messages=messages,      # ä¼ å…¥æ¶ˆæ¯åˆ—è¡¨
        max_tokens=1024         # é™åˆ¶æœ€å¤§tokenæ•°é‡
    )

    return response.choices[0].message.content.strip()

def parse_action(response: str) -> Dict:
    """
    è§£æLLMå“åº”ï¼Œæå–ç»“æ„åŒ–çš„å·¥å…·è°ƒç”¨æŒ‡ä»¤

    å‚æ•°:
        response: LLMçš„å“åº”æ–‡æœ¬

    è¿”å›:
        åŒ…å«å·¥å…·åç§°å’Œå‚æ•°çš„å­—å…¸
    """
    try:
        # ä»å“åº”ä¸­æå–actionä»£ç å—
        response = extract_markdown_block(response, "action")

        # å°†JSONå­—ç¬¦ä¸²è§£æä¸ºPythonå­—å…¸
        response_json = json.loads(response)

        # éªŒè¯å“åº”æ ¼å¼æ˜¯å¦æ­£ç¡®
        if "tool_name" in response_json and "args" in response_json:
            return response_json
        else:
            # å¦‚æœæ ¼å¼ä¸æ­£ç¡®ï¼Œè¿”å›é”™è¯¯ä¿¡æ¯
            return {"tool_name": "error", "args": {"message": "You must respond with a JSON tool invocation."}}
    except json.JSONDecodeError:
        # å¦‚æœJSONè§£æå¤±è´¥ï¼Œè¿”å›é”™è¯¯ä¿¡æ¯
        return {"tool_name": "error", "args": {"message": "Invalid JSON response. You must respond with a JSON tool invocation."}}

# ===== æ™ºèƒ½ä½“å¯ç”¨çš„å·¥å…·å‡½æ•° =====

def list_files() -> List[str]:
    """
    åˆ—å‡ºå½“å‰ç›®å½•ä¸­çš„æ‰€æœ‰æ–‡ä»¶

    è¿”å›:
        æ–‡ä»¶ååˆ—è¡¨
    """
    return os.listdir(".")

def read_file(file_name: str) -> str:
    """
    è¯»å–æŒ‡å®šæ–‡ä»¶çš„å†…å®¹

    å‚æ•°:
        file_name: è¦è¯»å–çš„æ–‡ä»¶å

    è¿”å›:
        æ–‡ä»¶å†…å®¹æˆ–é”™è¯¯ä¿¡æ¯
    """
    try:
        with open(file_name, "r") as file:
            return file.read()
    except FileNotFoundError:
        return f"Error: {file_name} not found."
    except Exception as e:
        return f"Error: {str(e)}"

# ===== æ™ºèƒ½ä½“ç³»ç»Ÿæç¤ºè¯å®šä¹‰ =====
# è¿™ä¸ªæç¤ºè¯å®šä¹‰äº†æ™ºèƒ½ä½“çš„è¡Œä¸ºè§„åˆ™å’Œå¯ç”¨å·¥å…·
agent_rules = [{
    "role": "system",
    "content": """
ä½ æ˜¯ä¸€ä¸ªAIæ™ºèƒ½ä½“ï¼Œå¯ä»¥é€šè¿‡ä½¿ç”¨å¯ç”¨å·¥å…·æ¥æ‰§è¡Œä»»åŠ¡ã€‚

å¯ç”¨å·¥å…·:

```json
{
    "list_files": {
        "description": "åˆ—å‡ºå½“å‰ç›®å½•ä¸­çš„æ‰€æœ‰æ–‡ä»¶ã€‚",
        "parameters": {}
    },
    "read_file": {
        "description": "è¯»å–æ–‡ä»¶çš„å†…å®¹ã€‚",
        "parameters": {
            "file_name": {
                "type": "string",
                "description": "è¦è¯»å–çš„æ–‡ä»¶åã€‚"
            }
        }
    },
    "terminate": {
        "description": "ç»“æŸæ™ºèƒ½ä½“å¾ªç¯å¹¶æä¾›ä»»åŠ¡æ‘˜è¦ã€‚",
        "parameters": {
            "message": {
                "type": "string",
                "description": "è¿”å›ç»™ç”¨æˆ·çš„æ‘˜è¦æ¶ˆæ¯ã€‚"
            }
        }
    }
}
```

å¦‚æœç”¨æˆ·è¯¢é—®æ–‡ä»¶ã€æ–‡æ¡£æˆ–å†…å®¹ï¼Œè¯·å…ˆåˆ—å‡ºæ–‡ä»¶ï¼Œç„¶åå†è¯»å–å®ƒä»¬ã€‚

å½“ä½ å®Œæˆä»»åŠ¡åï¼Œä½¿ç”¨"terminate"å·¥å…·ç»“æŸå¯¹è¯ï¼Œæˆ‘å°†å‘ç”¨æˆ·æä¾›ç»“æœã€‚

é‡è¦ï¼ï¼ï¼æ¯ä¸ªå“åº”éƒ½å¿…é¡»åŒ…å«ä¸€ä¸ªåŠ¨ä½œã€‚
ä½ å¿…é¡»å§‹ç»ˆæŒ‰ç…§ä»¥ä¸‹æ ¼å¼å“åº”ï¼š

<åœä¸‹æ¥é€æ­¥æ€è€ƒã€‚å‚æ•°æ˜ å°„åˆ°argsã€‚åœ¨è¿™é‡Œæ’å…¥ä½ é€æ­¥æ€è€ƒçš„ä¸°å¯Œæè¿°ã€‚>

```action
{
    "tool_name": "æ’å…¥å·¥å…·åç§°",
    "args": {...åœ¨è¿™é‡Œå¡«å…¥ä»»ä½•å¿…éœ€çš„å‚æ•°...}
}```
"""
}]

# ===== æ™ºèƒ½ä½“ä¸»å¾ªç¯åˆå§‹åŒ– =====
iterations = 0        # å½“å‰è¿­ä»£æ¬¡æ•°
max_iterations = 10   # æœ€å¤§è¿­ä»£æ¬¡æ•°ï¼Œé˜²æ­¢æ— é™å¾ªç¯

# è·å–ç”¨æˆ·ä»»åŠ¡
user_task = input("What would you like me to do? ")

# åˆå§‹åŒ–å¯¹è¯è®°å¿†ï¼ŒåŒ…å«ç”¨æˆ·çš„ä»»åŠ¡
memory = [{"role": "user", "content": user_task}]

# ===== æ™ºèƒ½ä½“ä¸»å¾ªç¯ =====
# è¿™æ˜¯æ™ºèƒ½ä½“çš„æ ¸å¿ƒå·¥ä½œå¾ªç¯ï¼ŒæŒç»­æ‰§è¡Œç›´åˆ°ä»»åŠ¡å®Œæˆæˆ–è¾¾åˆ°æœ€å¤§è¿­ä»£æ¬¡æ•°
while iterations < max_iterations:
    # 1. æ„å»ºæç¤ºï¼šå°†æ™ºèƒ½ä½“è§„åˆ™ä¸å¯¹è¯è®°å¿†ç»“åˆ
    prompt = agent_rules + memory

    # 2. è°ƒç”¨LLMç”Ÿæˆå“åº”
    print("Agent thinking...")
    response = generate_response(prompt)
    print(f"Agent response: {response}")

    # 3. è§£æå“åº”ä»¥ç¡®å®šè¦æ‰§è¡Œçš„åŠ¨ä½œ
    action = parse_action(response)
    result = "Action executed"  # é»˜è®¤ç»“æœ

    # 4. æ ¹æ®è§£æå‡ºçš„åŠ¨ä½œæ‰§è¡Œç›¸åº”çš„å·¥å…·å‡½æ•°
    if action["tool_name"] == "list_files":
        result = {"result": list_files()}
    elif action["tool_name"] == "read_file":
        result = {"result": read_file(action["args"]["file_name"])}
    elif action["tool_name"] == "error":
        result = {"error": action["args"]["message"]}
    elif action["tool_name"] == "terminate":
        print(action["args"]["message"])
        break  # ç»ˆæ­¢å¾ªç¯
    else:
        result = {"error": "Unknown action: " + action["tool_name"]}

    print(f"Action result: {result}")

    # 5. æ›´æ–°å¯¹è¯è®°å¿†ï¼Œæ·»åŠ æ™ºèƒ½ä½“å“åº”å’Œæ‰§è¡Œç»“æœ
    memory.extend([
        {"role": "assistant", "content": response},
        {"role": "user", "content": json.dumps(result)}
    ])

    # 6. æ£€æŸ¥ç»ˆæ­¢æ¡ä»¶
    if action["tool_name"] == "terminate":
        break

    iterations += 1  # å¢åŠ è¿­ä»£è®¡æ•°
````
