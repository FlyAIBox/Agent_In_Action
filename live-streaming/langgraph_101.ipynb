### ğŸ” ç¯å¢ƒä¿¡æ¯æ‰“å°

````python
# ğŸ” ç¯å¢ƒä¿¡æ¯æ£€æŸ¥è„šæœ¬
#
# æœ¬è„šæœ¬çš„ä½œç”¨ï¼š
# 1. å®‰è£… pandas åº“ç”¨äºæ•°æ®è¡¨æ ¼å±•ç¤º
# 2. æ£€æŸ¥ç³»ç»Ÿçš„å„é¡¹é…ç½®ä¿¡æ¯
# 3. ç”Ÿæˆè¯¦ç»†çš„ç¯å¢ƒæŠ¥å‘Šè¡¨æ ¼
#
# å¯¹äºåˆå­¦è€…æ¥è¯´ï¼Œè¿™ä¸ªæ­¥éª¤å¸®åŠ©ä½ ï¼š
# - äº†è§£å½“å‰è¿è¡Œç¯å¢ƒçš„ç¡¬ä»¶é…ç½®
# - ç¡®è®¤æ˜¯å¦æ»¡è¶³æ¨¡å‹è¿è¡Œçš„æœ€ä½è¦æ±‚
# - å­¦ä¹ å¦‚ä½•é€šè¿‡ä»£ç è·å–ç³»ç»Ÿä¿¡æ¯

# å®‰è£… pandas åº“ - ç”¨äºåˆ›å»ºå’Œå±•ç¤ºæ•°æ®è¡¨æ ¼
# pandas æ˜¯ Python ä¸­æœ€æµè¡Œçš„æ•°æ®å¤„ç†å’Œåˆ†æåº“
%pip install pandas==2.2.2 tabulate==0.9.0

import platform # å¯¼å…¥ platform æ¨¡å—ä»¥è·å–ç³»ç»Ÿä¿¡æ¯
import os # å¯¼å…¥ os æ¨¡å—ä»¥ä¸æ“ä½œç³»ç»Ÿäº¤äº’
import subprocess # å¯¼å…¥ subprocess æ¨¡å—ä»¥è¿è¡Œå¤–éƒ¨å‘½ä»¤
import pandas as pd # å¯¼å…¥ pandas æ¨¡å—ï¼Œé€šå¸¸ç”¨äºæ•°æ®å¤„ç†ï¼Œè¿™é‡Œç”¨äºåˆ›å»ºè¡¨æ ¼
import shutil # å¯¼å…¥ shutil æ¨¡å—ä»¥è·å–ç£ç›˜ç©ºé—´ä¿¡æ¯

# è·å– CPU ä¿¡æ¯çš„å‡½æ•°ï¼ŒåŒ…æ‹¬æ ¸å¿ƒæ•°é‡
def get_cpu_info():
    cpu_info = "" # åˆå§‹åŒ– CPU ä¿¡æ¯å­—ç¬¦ä¸²
    physical_cores = "N/A"
    logical_cores = "N/A"

    if platform.system() == "Windows": # å¦‚æœæ˜¯ Windows ç³»ç»Ÿ
        cpu_info = platform.processor() # ä½¿ç”¨ platform.processor() è·å– CPU ä¿¡æ¯
        try:
            # è·å– Windows ä¸Šçš„æ ¸å¿ƒæ•°é‡ (éœ€è¦ WMI)
            import wmi
            c = wmi.WMI()
            for proc in c.Win32_Processor():
                physical_cores = proc.NumberOfCores
                logical_cores = proc.NumberOfLogicalProcessors
        except:
            pass # å¦‚æœ WMI ä¸å¯ç”¨ï¼Œå¿½ç•¥é”™è¯¯

    elif platform.system() == "Darwin": # å¦‚æœæ˜¯ macOS ç³»ç»Ÿ
        # åœ¨ macOS ä¸Šä½¿ç”¨ sysctl å‘½ä»¤è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
        os.environ['PATH'] = os.environ['PATH'] + os.pathsep + '/usr/sbin' # æ›´æ–° PATH ç¯å¢ƒå˜é‡
        try:
            process_brand = subprocess.Popen(['sysctl', "machdep.cpu.brand_string"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_brand, stderr_brand = process_brand.communicate()
            cpu_info = stdout_brand.decode().split(': ')[1].strip() if stdout_brand else "Could not retrieve CPU info"

            process_physical = subprocess.Popen(['sysctl', "hw.physicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_physical, stderr_physical = process_physical.communicate()
            physical_cores = stdout_physical.decode().split(': ')[1].strip() if stdout_physical else "N/A"

            process_logical = subprocess.Popen(['sysctl', "hw.logicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_logical, stderr_logical = process_logical.communicate()
            logical_cores = stdout_logical.decode().split(': ')[1].strip() if stdout_logical else "N/A"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    else:  # Linux ç³»ç»Ÿ
        try:
            # åœ¨ Linux ä¸Šè¯»å– /proc/cpuinfo æ–‡ä»¶è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
            with open('/proc/cpuinfo') as f:
                physical_cores_count = 0
                logical_cores_count = 0
                cpu_info_lines = []
                for line in f:
                    if line.startswith('model name'): # æŸ¥æ‰¾ä»¥ 'model name'å¼€å¤´çš„è¡Œ
                        if not cpu_info: # åªè·å–ç¬¬ä¸€ä¸ª model name
                            cpu_info = line.split(': ')[1].strip()
                    elif line.startswith('cpu cores'): # æŸ¥æ‰¾ä»¥ 'cpu cores' å¼€å¤´çš„è¡Œ
                        physical_cores_count = int(line.split(': ')[1].strip())
                    elif line.startswith('processor'): # æŸ¥æ‰¾ä»¥ 'processor' å¼€å¤´çš„è¡Œ
                        logical_cores_count += 1
                physical_cores = str(physical_cores_count) if physical_cores_count > 0 else "N/A"
                logical_cores = str(logical_cores_count) if logical_cores_count > 0 else "N/A"
                if not cpu_info:
                     cpu_info = "Could not retrieve CPU info"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    return f"{cpu_info} ({physical_cores} physical cores, {logical_cores} logical cores)" # è¿”å› CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡


# è·å–å†…å­˜ä¿¡æ¯çš„å‡½æ•°
def get_memory_info():
    mem_info = "" # åˆå§‹åŒ–å†…å­˜ä¿¡æ¯å­—ç¬¦ä¸²
    if platform.system() == "Windows":
        # åœ¨ Windows ä¸Šä¸å®¹æ˜“é€šè¿‡æ ‡å‡†åº“è·å–ï¼Œéœ€è¦å¤–éƒ¨åº“æˆ– PowerShell
        mem_info = "Requires external tools on Windows" # è®¾ç½®æç¤ºä¿¡æ¯
    elif platform.system() == "Darwin": # å¦‚æœæ˜¯ macOS ç³»ç»Ÿ
        # åœ¨ macOS ä¸Šä½¿ç”¨ sysctl å‘½ä»¤è·å–å†…å­˜å¤§å°
        process = subprocess.Popen(['sysctl', "hw.memsize"], stdout=subprocess.PIPE, stderr=subprocess.PIPE) # è¿è¡Œ sysctl å‘½ä»¤
        stdout, stderr = process.communicate() # è·å–æ ‡å‡†è¾“å‡ºå’Œæ ‡å‡†é”™è¯¯
        mem_bytes = int(stdout.decode().split(': ')[1].strip()) # è§£æè¾“å‡ºï¼Œè·å–å†…å­˜å¤§å°ï¼ˆå­—èŠ‚ï¼‰
        mem_gb = mem_bytes / (1024**3) # è½¬æ¢ä¸º GB
        mem_info = f"{mem_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡º
    else:  # Linux ç³»ç»Ÿ
        try:
            # åœ¨ Linux ä¸Šè¯»å– /proc/meminfo æ–‡ä»¶è·å–å†…å­˜ä¿¡æ¯
            with open('/proc/meminfo') as f:
                total_mem_kb = 0
                available_mem_kb = 0
                for line in f:
                    if line.startswith('MemTotal'): # æŸ¥æ‰¾ä»¥ 'MemTotal' å¼€å¤´çš„è¡Œ
                        total_mem_kb = int(line.split(':')[1].strip().split()[0]) # è§£æè¡Œï¼Œè·å–æ€»å†…å­˜ï¼ˆKBï¼‰
                    elif line.startswith('MemAvailable'): # æŸ¥æ‰¾ä»¥ 'MemAvailable' å¼€å¤´çš„è¡Œ
                         available_mem_kb = int(line.split(':')[1].strip().split()[0]) # è§£æè¡Œï¼Œè·å–å¯ç”¨å†…å­˜ï¼ˆKBï¼‰

                if total_mem_kb > 0:
                    total_mem_gb = total_mem_kb / (1024**2) # è½¬æ¢ä¸º GB
                    mem_info = f"{total_mem_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡ºæ€»å†…å­˜
                    if available_mem_kb > 0:
                        available_mem_gb = available_mem_kb / (1024**2)
                        mem_info += f" (Available: {available_mem_gb:.2f} GB)" # æ·»åŠ å¯ç”¨å†…å­˜ä¿¡æ¯
                else:
                     mem_info = "Could not retrieve memory info" # å¦‚æœè¯»å–æ–‡ä»¶å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯

        except:
            mem_info = "Could not retrieve memory info" # å¦‚æœè¯»å–æ–‡ä»¶å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯
    return mem_info # è¿”å›å†…å­˜ä¿¡æ¯

# è·å– GPU ä¿¡æ¯çš„å‡½æ•°ï¼ŒåŒ…æ‹¬æ˜¾å­˜
def get_gpu_info():
    try:
        # å°è¯•ä½¿ç”¨ nvidia-smi è·å– NVIDIA GPU ä¿¡æ¯å’Œæ˜¾å­˜
        result = subprocess.run(['nvidia-smi', '--query-gpu=name,memory.total', '--format=csv,noheader'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            gpu_lines = result.stdout.strip().split('\n') # è§£æè¾“å‡ºï¼Œè·å– GPU åç§°å’Œæ˜¾å­˜
            gpu_info_list = []
            for line in gpu_lines:
                name, memory = line.split(', ')
                gpu_info_list.append(f"{name} ({memory})") # æ ¼å¼åŒ– GPU ä¿¡æ¯
            return ", ".join(gpu_info_list) if gpu_info_list else "NVIDIA GPU found, but info not listed" # è¿”å› GPU ä¿¡æ¯æˆ–æç¤ºä¿¡æ¯
        else:
             # å°è¯•ä½¿ç”¨ lshw è·å–å…¶ä»– GPU ä¿¡æ¯ (éœ€è¦å®‰è£… lshw)
            try:
                result_lshw = subprocess.run(['lshw', '-C', 'display'], capture_output=True, text=True)
                if result_lshw.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
                     # ç®€å•è§£æè¾“å‡ºä¸­çš„ product åç§°å’Œæ˜¾å­˜
                    gpu_info_lines = []
                    current_gpu = {}
                    for line in result_lshw.stdout.splitlines():
                        if 'product:' in line:
                             if current_gpu:
                                 gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")
                             current_gpu = {'product': line.split('product:')[1].strip()}
                        elif 'size:' in line and 'memory' in line:
                             current_gpu['memory'] = line.split('size:')[1].strip()

                    if current_gpu: # æ·»åŠ æœ€åä¸€ä¸ª GPU çš„ä¿¡æ¯
                        gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")

                    return ", ".join(gpu_info_lines) if gpu_info_lines else "GPU found (via lshw), but info not parsed" # å¦‚æœæ‰¾åˆ° GPU ä½†ä¿¡æ¯æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
                else:
                    return "No GPU found (checked nvidia-smi and lshw)" # å¦‚æœä¸¤ä¸ªå‘½ä»¤éƒ½æ‰¾ä¸åˆ° GPUï¼Œè®¾ç½®æç¤ºä¿¡æ¯
            except FileNotFoundError:
                 return "No GPU found (checked nvidia-smi, lshw not found)" # å¦‚æœæ‰¾ä¸åˆ° lshw å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "No GPU found (nvidia-smi not found)" # å¦‚æœæ‰¾ä¸åˆ° nvidia-smi å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯


# è·å– CUDA ç‰ˆæœ¬çš„å‡½æ•°
def get_cuda_version():
    try:
        # å°è¯•ä½¿ç”¨ nvcc --version è·å– CUDA ç‰ˆæœ¬
        result = subprocess.run(['nvcc', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            for line in result.stdout.splitlines():
                if 'release' in line: # æŸ¥æ‰¾åŒ…å« 'release' çš„è¡Œ
                    return line.split('release ')[1].split(',')[0] # è§£æè¡Œï¼Œæå–ç‰ˆæœ¬å·
        return "CUDA not found or version not parsed" # å¦‚æœæ‰¾ä¸åˆ° CUDA æˆ–ç‰ˆæœ¬æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "CUDA not found" # å¦‚æœæ‰¾ä¸åˆ° nvcc å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯

# è·å– Python ç‰ˆæœ¬çš„å‡½æ•°
def get_python_version():
    return platform.python_version() # è·å– Python ç‰ˆæœ¬

# è·å– Conda ç‰ˆæœ¬çš„å‡½æ•°
def get_conda_version():
    try:
        # å°è¯•ä½¿ç”¨ conda --version è·å– Conda ç‰ˆæœ¬
        result = subprocess.run(['conda', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            return result.stdout.strip() # è¿”å› Conda ç‰ˆæœ¬
        return "Conda not found or version not parsed" # å¦‚æœæ‰¾ä¸åˆ° Conda æˆ–ç‰ˆæœ¬æ— æ³•è§£æï¼Œè®¾ç½®æç¤ºä¿¡æ¯
    except FileNotFoundError:
        return "Conda not found" # å¦‚æœæ‰¾ä¸åˆ° conda å‘½ä»¤ï¼Œè®¾ç½®æç¤ºä¿¡æ¯

# è·å–ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯çš„å‡½æ•°
def get_disk_space():
    try:
        total, used, free = shutil.disk_usage("/") # è·å–æ ¹ç›®å½•çš„ç£ç›˜ä½¿ç”¨æƒ…å†µ
        total_gb = total / (1024**3) # è½¬æ¢ä¸º GB
        used_gb = used / (1024**3) # è½¬æ¢ä¸º GB
        free_gb = free / (1024**3) # è½¬æ¢ä¸º GB
        return f"Total: {total_gb:.2f} GB, Used: {used_gb:.2f} GB, Free: {free_gb:.2f} GB" # æ ¼å¼åŒ–è¾“å‡º
    except Exception as e:
        return f"Could not retrieve disk info: {e}" # å¦‚æœè·å–ä¿¡æ¯å‡ºé”™ï¼Œè®¾ç½®é”™è¯¯ä¿¡æ¯

# è·å–ç¯å¢ƒä¿¡æ¯
os_name = platform.system() # è·å–æ“ä½œç³»ç»Ÿåç§°
os_version = platform.release() # è·å–æ“ä½œç³»ç»Ÿç‰ˆæœ¬
if os_name == "Linux":
    try:
        # åœ¨ Linux ä¸Šå°è¯•è·å–å‘è¡Œç‰ˆå’Œç‰ˆæœ¬
        lsb_info = subprocess.run(['lsb_release', '-a'], capture_output=True, text=True)
        if lsb_info.returncode == 0: # å¦‚æœå‘½ä»¤æˆåŠŸæ‰§è¡Œ
            for line in lsb_info.stdout.splitlines():
                if 'Description:' in line: # æŸ¥æ‰¾åŒ…å« 'Description:' çš„è¡Œ
                    os_version = line.split('Description:')[1].strip() # æå–æè¿°ä¿¡æ¯ä½œä¸ºç‰ˆæœ¬
                    break # æ‰¾åˆ°åé€€å‡ºå¾ªç¯
                elif 'Release:' in line: # æŸ¥æ‰¾åŒ…å« 'Release:' çš„è¡Œ
                     os_version = line.split('Release:')[1].strip() # æå–ç‰ˆæœ¬å·
                     # å°è¯•è·å– codename
                     try:
                         codename_info = subprocess.run(['lsb_release', '-c'], capture_output=True, text=True)
                         if codename_info.returncode == 0:
                             os_version += f" ({codename_info.stdout.split(':')[1].strip()})" # å°† codename æ·»åŠ åˆ°ç‰ˆæœ¬ä¿¡æ¯ä¸­
                     except:
                         pass # å¦‚æœè·å– codename å¤±è´¥åˆ™å¿½ç•¥

    except FileNotFoundError:
        pass # lsb_release å¯èƒ½æœªå®‰è£…ï¼Œå¿½ç•¥é”™è¯¯

full_os_info = f"{os_name} {os_version}" # ç»„åˆå®Œæ•´çš„æ“ä½œç³»ç»Ÿä¿¡æ¯
cpu_info = get_cpu_info() # è°ƒç”¨å‡½æ•°è·å– CPU ä¿¡æ¯å’Œæ ¸å¿ƒæ•°é‡
memory_info = get_memory_info() # è°ƒç”¨å‡½æ•°è·å–å†…å­˜ä¿¡æ¯
gpu_info = get_gpu_info() # è°ƒç”¨å‡½æ•°è·å– GPU ä¿¡æ¯å’Œæ˜¾å­˜
cuda_version = get_cuda_version() # è°ƒç”¨å‡½æ•°è·å– CUDA ç‰ˆæœ¬
python_version = get_python_version() # è°ƒç”¨å‡½æ•°è·å– Python ç‰ˆæœ¬
conda_version = get_conda_version() # è°ƒç”¨å‡½æ•°è·å– Conda ç‰ˆæœ¬
disk_info = get_disk_space() # è°ƒç”¨å‡½æ•°è·å–ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯


# åˆ›å»ºç”¨äºå­˜å‚¨æ•°æ®çš„å­—å…¸
env_data = {
    "é¡¹ç›®": [ # é¡¹ç›®åç§°åˆ—è¡¨
        "æ“ä½œç³»ç»Ÿ",
        "CPU ä¿¡æ¯",
        "å†…å­˜ä¿¡æ¯",
        "GPU ä¿¡æ¯",
        "CUDA ä¿¡æ¯",
        "Python ç‰ˆæœ¬",
        "Conda ç‰ˆæœ¬",
        "ç‰©ç†ç£ç›˜ç©ºé—´" # æ·»åŠ ç‰©ç†ç£ç›˜ç©ºé—´
    ],
    "ä¿¡æ¯": [ # å¯¹åº”çš„ä¿¡æ¯åˆ—è¡¨
        full_os_info,
        cpu_info,
        memory_info,
        gpu_info,
        cuda_version,
        python_version,
        conda_version,
        disk_info # æ·»åŠ ç‰©ç†ç£ç›˜ç©ºé—´ä¿¡æ¯
    ]
}

# åˆ›å»ºä¸€ä¸ª pandas DataFrame
df = pd.DataFrame(env_data)

# æ‰“å°è¡¨æ ¼
print("### ç¯å¢ƒä¿¡æ¯") # æ‰“å°æ ‡é¢˜
print(df.to_markdown(index=False)) # å°† DataFrame è½¬æ¢ä¸º Markdown æ ¼å¼å¹¶æ‰“å°ï¼Œä¸åŒ…å«ç´¢å¼•
````

````python
import sqlite3
import requests
from sqlalchemy import create_engine
from sqlalchemy.pool import StaticPool

def show_graph(graph, xray=False):
    """å±•ç¤º LangGraph çš„ Mermaid æµç¨‹å›¾ï¼Œå¦‚æœå¤±è´¥åˆ™ä½¿ç”¨ ASCII å›¾è¡¨ä½œä¸ºåå¤‡ã€‚
    
    å‚æ•°:
        graph: æ‹¥æœ‰ get_graph() æ–¹æ³•çš„ LangGraph å¯¹è±¡
        xray: æ˜¯å¦å±•ç¤ºå›¾çš„å†…éƒ¨ç»“æ„ (ä¾‹å¦‚å­å›¾)
    """
    from IPython.display import Image
    try:
        return Image(graph.get_graph(xray=xray).draw_mermaid_png())
    except Exception as e:
        print(f"âš ï¸  å›¾ç‰‡æ¸²æŸ“å¤±è´¥: {e}")
        print("\nğŸ“Š æ”¹ä¸ºå±•ç¤º ASCII å›¾è¡¨:\n")
        ascii_diagram = graph.get_graph(xray=xray).draw_ascii()
        print(ascii_diagram)
        return None

def get_engine_for_chinook_db():
    """ä¸‹è½½ SQL æ–‡ä»¶ï¼Œå¡«å……å†…å­˜æ•°æ®åº“ï¼Œå¹¶åˆ›å»º SQLAlchemy å¼•æ“ã€‚"""
    url = "https://raw.githubusercontent.com/lerocha/chinook-database/master/ChinookDatabase/DataSources/Chinook_Sqlite.sql"
    response = requests.get(url)
    sql_script = response.text

    connection = sqlite3.connect(":memory:", check_same_thread=False)
    connection.executescript(sql_script)
    return create_engine(
        "sqlite://",
        creator=lambda: connection,
        poolclass=StaticPool,
        connect_args={"check_same_thread": False},
    )
````

- é€šè¿‡å‘åˆ—è¡¨è¿½åŠ  (append) æ¶ˆæ¯æ¥æ„å»ºå¤šè½®å¯¹è¯

## ç¬¬ 3 éƒ¨åˆ†: æ·»åŠ å·¥å…· - æ‰©å±• LLM èƒ½åŠ›

LLM æ“…é•¿å¤„ç†è¯­è¨€ï¼Œä½†å®ƒä»¬æ— æ³•è®¿é—®å¤–éƒ¨æ•°æ®æˆ–æ‰§è¡ŒåŠ¨ä½œã€‚**Tools (å·¥å…·)** æ‰©å±•äº†å®ƒä»¬çš„èƒ½åŠ›ã€‚ä½ å¯ä»¥ç»™ LLM æä¾›ä¸€ä¸ªå·¥å…·åˆ—è¡¨ï¼Œå½“å®ƒéœ€è¦æ—¶ï¼Œå®ƒä¼šæŒ‡å®šè¦è°ƒç”¨å“ªä¸ªå·¥å…·ã€‚ä½ çš„å·¥ä½œæ˜¯æ‰§è¡Œè¿™äº›å·¥å…·ï¼Œå¹¶å°†ç»“æœåé¦ˆç»™ LLMï¼Œä»¥ä¾¿å®ƒå†³å®šä¸‹ä¸€æ­¥åšä»€ä¹ˆã€‚

ä½ å¯ä»¥é€šè¿‡ç¼–å†™ä¸€ä¸ªå¸¦æœ‰æ¸…æ™°æè¿°çš„ Python å‡½æ•°æ¥åˆ›å»ºä¸€ä¸ªå·¥å…·ã€‚LangChain çš„ `@tool` è£…é¥°å™¨ä¼šè‡ªåŠ¨å¤„ç†å‡½æ•°ä¿¡æ¯çš„æ ¼å¼åŒ–ï¼Œä½¿å…¶ç¬¦åˆ LLM çš„è¦æ±‚ã€‚

è®©æˆ‘ä»¬åˆ›å»ºä¸€äº›ç®€å•çš„å·¥å…·ï¼š

````python
import requests
import json

from langchain_core.tools import tool


# ä¸€ä¸ªåŸºæœ¬çš„ç¡¬ç¼–ç å·¥å…·
@tool
def search_movies(genre: str) -> str:
    """æ ¹æ®ç±»å‹æœç´¢ç”µå½±æ¨èã€‚""" # å‡½æ•°çš„ docstring ä¼šä½œä¸ºå·¥å…·æè¿°æä¾›ç»™ LLM
    # åœ¨çœŸå®åº”ç”¨ä¸­ï¼Œè¿™å¯èƒ½ä¼šæŸ¥è¯¢ç”µå½±æ•°æ®åº“
    movies = {
    "ç§‘å¹»": "æ²™ä¸˜, æ˜Ÿé™…ç©¿è¶Š, é“¶ç¿¼æ€æ‰‹2049",
    "å–œå‰§": "å¸ƒè¾¾ä½©æ–¯å¤§é¥­åº—, è¶…çº§å, åˆ©åˆƒå‡ºé˜",
    "åŠ¨ä½œ": "ç–¯ç‹‚çš„éº¦å…‹æ–¯ï¼šç‹‚æš´ä¹‹è·¯, ç–¾é€Ÿè¿½æ€, ç¢Ÿä¸­è°"
    }
    return movies.get(genre.lower(), "No movies found for that genre")

# ä¸€ä¸ªæ›´çœŸå®çš„å·¥å…·ï¼Œè°ƒç”¨å¤–éƒ¨ API
@tool
def get_weather(latitude: float, longitude: float) -> str:
    """è·å–æŒ‡å®šåæ ‡çš„å½“å‰æ¸©åº¦ (åæ°åº¦) å’Œå¤©æ°”ä»£ç ã€‚

    Args:
        latitude: çº¬åº¦åæ ‡
        longitude: ç»åº¦åæ ‡

    Returns:
        åŒ…å« temperature_fahrenheit å’Œ weather_code çš„ JSON å­—ç¬¦ä¸² (ä¸åŒ…å«ä»£ç è§£é‡Šï¼Œç•™ç»™æ¨¡å‹å¤„ç†)
    """
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current": "temperature_2m,weather_code",
        "temperature_unit": "fahrenheit"
    }

    weather = requests.get(url, params=params).json()["current"]
    temperature = weather["temperature_2m"]
    weather_code = weather["weather_code"]
    result = {
        "temperature_fahrenheit": temperature,
        "weather_code": weather_code
    }

    return json.dumps(result)



# ç›´æ¥æµ‹è¯•å·¥å…·ï¼Œä¾‹å¦‚æŸ¥è¯¢æ—§é‡‘å±±çš„åæ ‡
print(get_weather.invoke({"latitude": 31.23, "longitude": 121.47}))
# æŸ¥è¯¢åŒ—äº¬çš„å¤©æ°”
print(get_weather.invoke({"latitude": 39.90, "longitude": 116.40}))
````

### å·¥å…·è°ƒç”¨ (Tool Calling / Function Calling)

ç°åœ¨ï¼Œè®©æˆ‘ä»¬ä½¿ç”¨ `.bind_tools()` å°†è¿™äº›å·¥å…·æä¾›ç»™æ¨¡å‹ï¼š

````python
# å°†å·¥å…·ç»‘å®šåˆ°æ¨¡å‹
tools = [get_weather, search_movies]
model_with_tools = model.bind_tools(tools)

message = "æ­å·çš„å¤©æ°”æ€ä¹ˆæ ·ï¼Ÿï¼ˆæ­å·çš„åæ ‡å¤§çº¦æ˜¯åŒ—çº¬ 30.27Â°ï¼Œä¸œç» 120.15Â°ï¼‰"

# æ¨¡å‹ç°åœ¨å¯ä»¥å†³å®šæ˜¯å¦è°ƒç”¨å·¥å…·
response = model_with_tools.invoke(message)

# æ£€æŸ¥æ¨¡å‹æ˜¯å¦æƒ³è¦è°ƒç”¨å·¥å…·
# tool_calls å±æ€§åŒ…å«äº†æ¨¡å‹ç”Ÿæˆçš„ç»“æ„åŒ–å·¥å…·è°ƒç”¨è¯·æ±‚
print("Tool calls:", response.tool_calls)
````

æ¨¡å‹è¿”å›ä¸€ä¸ª **tool call (å·¥å…·è°ƒç”¨)** è¯·æ±‚ï¼Œå…¶ä¸­åŒ…å«ï¼š
- `name`: è¦è°ƒç”¨çš„å·¥å…·åç§°
- `args`: ä¼ é€’ç»™å·¥å…·çš„å‚æ•°
- `id`: ç”¨äºè·Ÿè¸ªçš„å”¯ä¸€æ ‡è¯†ç¬¦

è®©æˆ‘ä»¬æ‰§è¡Œå·¥å…·å¹¶ç»§ç»­å¯¹è¯ï¼š

````python
from langchain.messages import ToolMessage

# æ‰§è¡Œå·¥å…·è°ƒç”¨
if response.tool_calls:
    tool_call = response.tool_calls[0]

    # è°ƒç”¨å®é™…çš„å·¥å…·å‡½æ•°
    if tool_call["name"] == "get_weather":
        result = get_weather.invoke(tool_call["args"])

    elif tool_call["name"] == "search_movies":
        result = search_movies.invoke(tool_call["args"])

    # åˆ›å»ºä¸€ä¸ª ToolMessage æ¥å°è£…ç»“æœ
    # æ³¨æ„ï¼šå¿…é¡»ä¼ å…¥ tool_call_idï¼Œä»¥ä¾¿æ¨¡å‹çŸ¥é“è¿™ä¸ªç»“æœå¯¹åº”å“ªä¸ªè°ƒç”¨
    tool_message = ToolMessage(
        content=result,
        tool_call_id=tool_call["id"]
    )

    # å°†å·¥å…·æ‰§è¡Œç»“æœåé¦ˆç»™æ¨¡å‹ï¼Œç»§ç»­ç”Ÿæˆæœ€ç»ˆå›å¤
    final_response = model_with_tools.invoke([
        HumanMessage(content=message),
        response,    # æ¨¡å‹çš„åŸå§‹å›å¤ï¼ˆåŒ…å«å·¥å…·è°ƒç”¨è¯·æ±‚ï¼‰
        tool_message # æˆ‘ä»¬æ‰§è¡Œå·¥å…·åå¾—åˆ°çš„ç»“æœ
    ])

    final_response.pretty_print()
````

### æ ¸å¿ƒè¦ç‚¹ï¼š
- å·¥å…·æ˜¯ä½¿ç”¨ `@tool` è£…é¥°çš„ Python å‡½æ•°
- è‰¯å¥½çš„æ–‡æ¡£æ³¨é‡Š (Docstrings) èƒ½å¸®åŠ©æ¨¡å‹åˆ¤æ–­ä½•æ—¶ä½¿ç”¨è¯¥å·¥å…·
- å·¥å…·è°ƒç”¨æµç¨‹ï¼šæ¨¡å‹è¯·æ±‚è°ƒç”¨å·¥å…· â†’ ä»£ç æ‰§è¡Œå·¥å…· â†’ è¿”å›ç»“æœç»™æ¨¡å‹ â†’ æ¨¡å‹ç»¼åˆä¿¡æ¯ç”Ÿæˆæœ€ç»ˆå›å¤

## ç¬¬ 4 éƒ¨åˆ†: ä½¿ç”¨ `create_agent()` æ„å»ºä½ çš„ç¬¬ä¸€ä¸ª Agent

æ‰‹åŠ¨å¤„ç† LLM è°ƒç”¨ã€æ£€æŸ¥å·¥å…·è°ƒç”¨ã€æ‰§è¡Œå·¥å…·ã€å†å›ä¼ ç»“æœçš„å¾ªç¯éå¸¸ç¹çä¸”ä¸çµæ´»ã€‚ç›¸åï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨ **Agent** æ¥è‡ªåŠ¨è¿è¡Œè¿™ä¸ªå¾ªç¯ï¼š
1. æ¨¡å‹å†³å®šè°ƒç”¨å“ªä¸ªå·¥å…·ï¼ˆå¦‚æœæœ‰ï¼‰
2. å·¥å…·è¢«æ‰§è¡Œ
3. ç»“æœè¿”å›ç»™æ¨¡å‹
4. é‡å¤ä¸Šè¿°æ­¥éª¤ç›´åˆ°ä»»åŠ¡å®Œæˆ

LangChain æä¾›äº† `create_agent()` é«˜çº§ API - **ä»…éœ€çº¦ 10 è¡Œä»£ç å³å¯æ„å»ºä¸€ä¸ª Agentï¼**
è¿™ä¸ªé¢„ç½®çš„ Agent ä¼šè‡ªåŠ¨å¤„ç†ä¸Šè¿°å¾ªç¯ â€”â€” ä½ åªéœ€è¦æŒ‡å®šç³»ç»Ÿæç¤ºè¯ (System Prompt) å’Œå·¥å…·åˆ—è¡¨ã€‚

````python
from langchain.agents import create_agent

# åˆ›å»ºä¸€ä¸ªå¸¦æœ‰å·¥å…·çš„ Agent
# è¿™æ˜¯ä¸€ä¸ªé«˜çº§ APIï¼Œåº•å±‚ä½¿ç”¨äº† LangGraph
agent = create_agent(
    model=model,
    tools=[get_weather, search_movies],
    system_prompt="ä½ æ˜¯ä¸€ä¸ªä¹äºåŠ©äººçš„åŠ©æ‰‹ï¼Œå¯ä»¥æŸ¥è¯¢å¤©æ°”å¹¶æ¨èç”µå½±ã€‚"
)

# ä½¿ç”¨ Agent
result = agent.invoke({
    "messages": [HumanMessage(content="å¹¿å·çš„å¤©æ°”æ€ä¹ˆæ ·ï¼Ÿ(åŒ—çº¬ 23.13Â°ï¼Œä¸œç» 113.26Â°) å¦å¤–æ¨èä¸€äº›ç§‘å¹»ç”µå½±ã€‚")]
})

# æ‰“å°æœ€ç»ˆå“åº”
for message in result["messages"]:
    message.pretty_print()
````

### æ ¸å¿ƒè¦ç‚¹ï¼š

- Agent è‡ªåŠ¨å¤„ç†äº† æ¨ç† (Reasoning) â†’ è¡ŒåŠ¨ (Action) â†’ è§‚å¯Ÿ (Observation) çš„å¾ªç¯
- **LangGraph 1.0 ç‰¹æ€§**ï¼šå®ƒå»ºç«‹åœ¨ LangGraph ä¹‹ä¸Šï¼Œå¤©ç”Ÿå…·å¤‡ç”Ÿäº§çº§ç‰¹æ€§ï¼ˆæŒä¹…åŒ–ã€æµå¼ä¼ è¾“ã€äººæœºäº¤äº’ï¼‰

## ç¬¬ 5 éƒ¨åˆ†: æ·»åŠ è®°å¿† (Memory) & çŠ¶æ€ (State)

ç›®å‰ï¼Œæ¯æ¬¡ Agent è°ƒç”¨éƒ½æ˜¯ç‹¬ç«‹çš„ã€‚è®©æˆ‘ä»¬æ·»åŠ  **Memory (è®°å¿†)**ï¼Œä»¥ä¾¿ Agent èƒ½å¤Ÿåœ¨å¤šæ¬¡äº¤äº’ä¸­ä¿æŒä¸Šä¸‹æ–‡ã€‚

LangGraph ä½¿ç”¨ **Checkpointers (æ£€æŸ¥ç‚¹ä¿å­˜å™¨)** æ¥ä¿å­˜å’Œæ¢å¤çŠ¶æ€ï¼š

````python
from langgraph.checkpoint.memory import MemorySaver
from langsmith import uuid7

# åˆ›å»ºä¸€ä¸ªå†…å­˜æ£€æŸ¥ç‚¹ä¿å­˜å™¨ (åœ¨ç”Ÿäº§ç¯å¢ƒä¸­ï¼Œé€šå¸¸ä½¿ç”¨ PostgresSaver ç­‰)
checkpointer = MemorySaver()

# åˆ›å»ºå¸¦æœ‰è®°å¿†çš„ Agent
agent_with_memory = create_agent(
    model=model,
    tools=[get_weather, search_movies],
    system_prompt="ä½ æ˜¯ä¸€ä¸ªä¹äºåŠ©äººçš„åŠ©æ‰‹ã€‚",
    checkpointer=checkpointer # ä¼ å…¥ checkpointer å¯ç”¨æŒä¹…åŒ–
)

# ä¸ºæ­¤å¯¹è¯åˆ›å»ºä¸€ä¸ªå”¯ä¸€çš„çº¿ç¨‹ ID (Thread ID)
config = {"configurable": {"thread_id": uuid7()}}

# ç¬¬ä¸€æ¬¡äº¤äº’
result1 = agent_with_memory.invoke(
    {"messages": [{"role": "user", "content": "æˆ‘å« FLYï¼Œæˆ‘å–œæ¬¢ç§‘å¹»ç”µå½±ã€‚"}]},
    config=config
)

print("Response 1:", result1["messages"][-1].content)

# ç¬¬äºŒæ¬¡äº¤äº’ - Agent è®°å¾—æˆ‘ä»¬ï¼
result2 = agent_with_memory.invoke(
    {"messages": [{"role": "user", "content": "æˆ‘å«ä»€ä¹ˆåå­—ï¼Ÿæˆ‘å–œæ¬¢ä»€ä¹ˆç”µå½±ï¼Ÿ"}]},
    config=config
)
print("\nResponse 2:", result2["messages"][-1].content)
````

### ç†è§£çŠ¶æ€ (State) & çº¿ç¨‹ (Threads)

- **State (çŠ¶æ€)**: Agent çš„â€œè®°å¿†â€ â€”â€” åŒ…å«äº†æ¶ˆæ¯å†å²å’Œä»»ä½•è‡ªå®šä¹‰æ•°æ®ã€‚
- **Thread (çº¿ç¨‹)**: ç”± `thread_id` æ ‡è¯†çš„ä¸€ä¸ªå¯¹è¯ä¼šè¯ã€‚
- **Checkpointer**: åœ¨æ¯ä¸€æ­¥ä¹‹åä¿å­˜çŠ¶æ€ï¼Œå®ç°äº†è®°å¿†åŠŸèƒ½å’Œé”™è¯¯æ¢å¤èƒ½åŠ›ã€‚

æ¯ä¸ªçº¿ç¨‹éƒ½æ˜¯ç‹¬ç«‹çš„ï¼š

````python
# æ–°çš„çº¿ç¨‹ - Agent ä¸ä¼šè®°å¾— FLY
new_config_with_new_thread_id = {"configurable": {"thread_id": uuid7()}}

result3 = agent_with_memory.invoke(
    {"messages": [{"role": "user", "content": "æˆ‘å«ä»€ä¹ˆåå­—ï¼Ÿ"}]},
    config=new_config_with_new_thread_id
)
print("New thread response:", result3["messages"][-1].content)
````

### æ ¸å¿ƒè¦ç‚¹ï¼š
- Checkpointers å®ç°äº†è·¨äº¤äº’çš„è®°å¿†
- Thread ID (çº¿ç¨‹ ID) ç”¨äºéš”ç¦»ä¸åŒçš„å¯¹è¯
- çŠ¶æ€ä¼šè‡ªåŠ¨æŒä¹…åŒ– â€”â€” æ— éœ€æ‰‹åŠ¨ç®¡ç†çŠ¶æ€ä¿å­˜ï¼

## ç¬¬ 6 éƒ¨åˆ†: æµå¼ä¼ è¾“ (Streaming) æå‡ç”¨æˆ·ä½“éªŒ

LLM å“åº”å¯èƒ½éœ€è¦ä¸€äº›æ—¶é—´ã€‚**Streaming (æµå¼ä¼ è¾“)** å¯ä»¥å®æ—¶æ˜¾ç¤ºè¿›åº¦ï¼Œæå¤§åœ°æå‡ç”¨æˆ·ä½“éªŒã€‚

LangGraph æ”¯æŒå¤šç§æµå¼æ¨¡å¼ï¼š

### æµå¼ä¼ è¾“ Agent æ­¥éª¤ (Streaming Steps)

````python
# ä½¿ç”¨ stream_mode="updates" æµå¼ä¼ è¾“ Agent çš„è¿›åº¦
print("Streaming agent steps:\n")

for chunk in agent.stream(
    {"messages": [{"role": "user", "content": "æˆéƒ½çš„å¤©æ°”æ€ä¹ˆæ ·ï¼Ÿ(åŒ—çº¬ 30.66Â°ï¼Œä¸œç» 104.07Â°) "}]},
    stream_mode="updates"
):
    # chunk æ˜¯è¿™ä¸€æ­¥çš„çŠ¶æ€æ›´æ–°
    for node_name, data in chunk.items():
        print(f"Step: {node_name}")
        if "messages" in data:
            message = data["messages"][-1]
            if hasattr(message, 'tool_calls') and message.tool_calls:
                print(f"   Tool call: {message.tool_calls[0]['name']}")
            elif hasattr(message, 'content'):
                content_preview = message.content[:100]
                print(f"   Content: {content_preview}...")
        print()
````

### æµå¼ä¼ è¾“ LLM Token

ä¸ºäº†è·å¾—ç±»ä¼¼ ChatGPT çš„ä½“éªŒï¼Œæˆ‘ä»¬å¯ä»¥æµå¼ä¼ è¾“ç”Ÿæˆçš„ Tokenï¼š

````python
# ä½¿ç”¨ stream_mode="messages" æµå¼ä¼ è¾“è¯¦ç»†æ¶ˆæ¯å†…å®¹ï¼ˆåŒ…æ‹¬ Tokenï¼‰
print("Streaming tokens:\n")

for token, metadata in agent.stream(
    {"messages": [{"role": "user", "content": "ç”¨ä¸€å¥è¯ä»‹ç»ä¸€ä¸‹ LangGraphã€‚"}]},
    stream_mode="messages"
):
    # ä»…æ‰“å°æ¥è‡ªæ¨¡å‹èŠ‚ç‚¹çš„ç”¨äºå±•ç¤ºçš„å†…å®¹
    if metadata.get('langgraph_node') == 'model':
        # è·å–æ–‡æœ¬å†…å®¹å—
        for block in token.content_blocks:
            if block.get('type') == 'text' and block.get('text'):
                print(block['text'], end='', flush=True)

print("\n")  # ç»“å°¾æ¢è¡Œ
````

### æ ¸å¿ƒè¦ç‚¹ï¼š
- `stream_mode="updates"` - æŸ¥çœ‹ Agent çš„æ¯ä¸€æ­¥æ“ä½œï¼ˆéå¸¸é€‚åˆè°ƒè¯•å’Œå±•ç¤ºæ€è€ƒè¿‡ç¨‹ï¼‰
- `stream_mode="messages"` - æµå¼ä¼ è¾“ LLM Tokenï¼ˆé€‚åˆå®ç°æ‰“å­—æœºæ•ˆæœï¼‰
- ä¸è®ºæ˜¯æ™®é€šè°ƒç”¨è¿˜æ˜¯æµå¼è°ƒç”¨ï¼Œå†™æ³•å‡ ä¹ä¸€è‡´ï¼Œæ— éœ€é¢å¤–é…ç½®ï¼

## ç¬¬ 7 éƒ¨åˆ†: ç»¼åˆå®æˆ˜ - ä¸€ä¸ªå®ç”¨çš„ä¾‹å­

è®©æˆ‘ä»¬æ•´åˆæ‰€å­¦çŸ¥è¯†ï¼Œæ„å»ºä¸€ä¸ªæ›´çœŸå®çš„ Agentï¼š

## ä»é›¶å¼€å§‹æ„å»º ReAct Agent

ä¸ºäº†ç†è§£ LangGraph çš„åŸè¯­æ˜¯å¦‚ä½•å·¥ä½œçš„ï¼Œè®©æˆ‘ä»¬ä¸ä½¿ç”¨ `create_agent()`ï¼Œè€Œæ˜¯ä»å¤´é‡æ„ä¸Šé¢çš„ Agentã€‚

### State (çŠ¶æ€)

æˆ‘ä»¬çš„èŠå¤©æœºå™¨äººå¾ˆç®€å•â€”â€”ä¸»è¦æ˜¯ LLM å’Œä¸€äº›å·¥å…·â€”â€”æ‰€ä»¥æˆ‘ä»¬åªéœ€è¦åœ¨çŠ¶æ€ä¸­ä¿ç•™ä¸€ä¸ªæ¶ˆæ¯åˆ—è¡¨ (`messages`)ï¼Œéšç€å¯¹è¯è¿›è¡Œï¼Œå®ƒä¼šä¸æ–­å¢é•¿ã€‚

````python
from typing_extensions import TypedDict
from typing import Annotated, List
from langgraph.graph.message import AnyMessage, add_messages

class State(TypedDict):
    # messages åŒ…å«å®Œæ•´çš„å¯¹è¯å†å²
    # Annotated[List, add_messages] å‘Šè¯‰ LangGraphï¼šå½“æœ‰æ–°æ¶ˆæ¯æ—¶ï¼Œä¸è¦è¦†ç›–ï¼Œè€Œæ˜¯â€œè¿½åŠ â€åˆ°åˆ—è¡¨ä¸­
    messages: Annotated[List[AnyMessage], add_messages]
````

### Nodes (èŠ‚ç‚¹)

æˆ‘ä»¬å¯ä»¥é‡ç”¨ä¸Šé¢å®šä¹‰çš„å·¥å…·ã€‚æˆ‘ä»¬éœ€è¦ä¸€ä¸ªèŠ‚ç‚¹æ¥å¤„ç†å·¥å…·è°ƒç”¨å¹¶å°†ç»“æœæ ¼å¼åŒ–ä¸º `ToolMessage`ã€‚LangGraph æä¾›äº†ä¸€ä¸ªé¢„æ„å»ºçš„ `ToolNode` æ¥å¸®æˆ‘ä»¬åšè¿™ä»¶äº‹ï¼Œæ— éœ€è‡ªå·±ç¼–å†™ã€‚

````python
from langgraph.prebuilt import ToolNode

# å·¥å…·èŠ‚ç‚¹
tools = [search_movies, get_weather]
tool_node = ToolNode(tools)
````

ç°åœ¨ï¼Œè®©æˆ‘ä»¬ä¸ºæ ¸å¿ƒçš„ Assistant (åŠ©æ‰‹) åˆ›å»ºä¸€ä¸ªèŠ‚ç‚¹ã€‚

````python
def assistant(state: State):
    # åˆ›å»ºå‘é€ç»™ LLM çš„æ¶ˆæ¯åˆ—è¡¨ï¼Œä»¥å›ºå®šçš„ System Prompt å¼€å¤´
    system_prompt = "You are a helpful assistant that can check weather and recommend movies."
    all_messages = [SystemMessage(system_prompt)] + state["messages"]
    
    # è°ƒç”¨ LLM
    response = model_with_tools.invoke(all_messages)
    
    # æ›´æ–°çŠ¶æ€ï¼Œè¿”å› LLM çš„å“åº”
    # ç”±äºæˆ‘ä»¬åœ¨ State ä¸­ä½¿ç”¨äº† add_messagesï¼Œè¿™é‡Œçš„ response ä¼šè¢«è¿½åŠ åˆ° messages åˆ—è¡¨æœ«å°¾
    return {"messages": [response]}
````

ç°åœ¨ï¼Œæˆ‘ä»¬éœ€è¦å®šä¹‰èŠ‚ç‚¹ä¹‹é—´çš„æ§åˆ¶æµï¼Œè¿™å°±è¦ç”¨åˆ° **Edges (è¾¹)**ã€‚

**Edges è¿æ¥èŠ‚ç‚¹ï¼Œå®šä¹‰äº† Graph çš„æµå‘ã€‚**
* **Normal edges (æ™®é€šè¾¹)** æ˜¯ç¡®å®šæ€§çš„ï¼Œæ€»æ˜¯ä»ä¸€ä¸ªèŠ‚ç‚¹æµå‘æŒ‡å®šçš„ä¸‹ä¸€ä¸ªèŠ‚ç‚¹ã€‚
* **Conditional edges (æ¡ä»¶è¾¹)** ç”¨äºåŠ¨æ€è·¯ç”±ï¼Œå®ƒæ˜¯ä¸€ä¸ªå‡½æ•°ï¼Œæ ¹æ®é€»è¾‘è¿”å›ä¸‹ä¸€ä¸ªè¦è®¿é—®çš„èŠ‚ç‚¹ã€‚

åœ¨è¿™ä¸ªä¾‹å­ä¸­ï¼Œæˆ‘ä»¬éœ€è¦ä¸€ä¸ªä» Assistant å‡ºæ¥çš„ **æ¡ä»¶è¾¹**ï¼Œå®ƒå†³å®šäº†ï¼š
- æ˜¯è°ƒç”¨å·¥å…· (Invoke tools)ï¼Ÿ
- è¿˜æ˜¯ç»“æŸå¯¹è¯ (End)ï¼Ÿ

````python
# å®šä¹‰æ¡ä»¶è¾¹é€»è¾‘ï¼šå†³å®šæ˜¯ç»§ç»­è¿˜æ˜¯ç»“æŸ
def should_continue(state: State):
    messages = state["messages"]
    last_message = messages[-1]
    
    # å¦‚æœ LLM æƒ³è¦è°ƒç”¨å·¥å…·ï¼Œæˆ‘ä»¬åº”è¯¥è½¬å»æ‰§è¡Œå·¥å…·
    if last_message.tool_calls:
        return "continue"
        
    # å¦‚æœ LLM è¿”å›äº†æ–‡æœ¬å“åº”è€Œä¸æ˜¯å·¥å…·è°ƒç”¨ï¼Œæˆ‘ä»¬ç»“æŸ
    else:
        return "end"
````

#### ç¼–è¯‘ Graph! (Compile Graph)

ç°åœ¨æˆ‘ä»¬å·²ç»å®šä¹‰äº† Stateã€Nodes å’Œ Conditional Edgeï¼Œè®©æˆ‘ä»¬æŠŠå®ƒä»¬ç»„è£…èµ·æ¥ï¼Œæ„å»ºæˆ‘ä»¬çš„ ReAct Agentï¼

````python
from langgraph.graph import StateGraph, START, END

# 1. ä½¿ç”¨æˆ‘ä»¬çš„ State ç±»å‹åˆå§‹åŒ– Graph Builder
builder = StateGraph(State)

# 2. æ·»åŠ èŠ‚ç‚¹
builder.add_node("assistant", assistant)
builder.add_node("tool_node", tool_node)


# 3. æ·»åŠ è¾¹
# é¦–å…ˆï¼Œå®šä¹‰èµ·å§‹è¾¹ã€‚è¯·æ±‚æ€»æ˜¯é¦–å…ˆè¿›å…¥ assistant èŠ‚ç‚¹ã€‚
builder.add_edge(START, "assistant")

# åœ¨è°ƒç”¨ LLM åæ·»åŠ æ¡ä»¶è¾¹
builder.add_conditional_edges(
    "assistant",       # ä» assistant èŠ‚ç‚¹å‡ºå‘
    should_continue,   # å†³ç­–å‡½æ•°
    {
        # å¦‚æœè¿”å› `continue`ï¼Œåˆ™å» `tool_node`
        "continue": "tool_node",
        # å¦åˆ™ç»“æŸ
        "end": END,
    },
)

# åœ¨å·¥å…·æ‰§è¡Œå®Œåï¼Œæ€»æ˜¯å›åˆ° assistantï¼ˆè®© LLM çœ‹åˆ°ç»“æœå¹¶å†³å®šä¸‹ä¸€æ­¥ï¼‰
builder.add_edge("tool_node", "assistant")

# 4. ç¼–è¯‘ Graph
agent = builder.compile(name="agent")
agent
````

æˆ‘ä»¬å·²ç»æˆåŠŸåœ°ä»é›¶å¼€å§‹é‡å»ºäº†ä¸€ä¸ª ReAct Agentï¼Œå…¶åŠŸèƒ½ä¸æˆ‘ä»¬ä¹‹å‰ç”¨ `create_agent` åˆ›å»ºçš„å®Œå…¨ç›¸åŒã€‚æ¥è¯•è¯•çœ‹å§ï¼

````python
from langchain.messages import HumanMessage

question = "ä»Šå¤©ä¸Šæµ·çš„å¤©æ°”æ€ä¹ˆæ ·ï¼ˆ31.23Â° N, 121.47Â° Eï¼‰ï¼Œæœ‰å“ªäº›å¥½çš„ç§‘å¹»ç”µå½±?"

# è¿™æ¬¡ç›´æ¥é€šè¿‡ Graph å¯¹è±¡è°ƒç”¨ invoke
result = agent.invoke({"messages": HumanMessage(content=question)})

for message in result["messages"]:
    message.pretty_print()
````
